```python
import tensorflow as tf
import numpy as np
from keras.preprocessing import image
from pathlib import Path
```

```python
MODEL_PATH = "outputs/dataset-224/best_model_fold0.keras"
IMG_PATH = Path.cwd().parent / '1-dataset-processing' / 'data' / 'dataset-article' / 'dataset-224' / 'with-stone' / 'imagestone300-patch5.png'
IMG_SIZE = (224, 224)
```

```python
model = tf.keras.models.load_model(MODEL_PATH)
print("Modelo carregado:", MODEL_PATH)
```

```python
img = image.load_img(IMG_PATH, target_size=IMG_SIZE)
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # adicionar batch
img_array = img_array / 255.0  # normalizar como no treino
```

```python
pred = model.predict(img_array)
prob = float(pred[0][0])  # valor entre 0 e 1
```

```python
if prob >= 0.5:
    print(f"A imagem {Path(IMG_PATH).name} foi classificada como: COM PEDRA (1). Probabilidade={prob:.4f}")
else:
    print(f"A imagem {Path(IMG_PATH).name} foi classificada como: SEM PEDRA (0). Probabilidade={prob:.4f}")
```
